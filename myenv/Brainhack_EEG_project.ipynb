```python
# Define the pattern for file search (e.g., '*.txt' for text files)
file_pattern = 'sub-*_task-Rest_run-01_eeg.fdt'

# Get the current directory
current_dir = os.getcwd()
print('The current directory is', current_dir)

# Initialize file count
file_count = 0

# Iterate over subdirectories recursively
for root, dirs, files in os.walk(current_dir):
    # Check if the current directory contains the files we need
    matching_files = [f for f in files if glob.fnmatch.fnmatch(f, file_pattern)]
    
    # Process matching files
    for fdt_file in matching_files:
        # Construct the full path to the .fdt file
        fdt_file_path = os.path.join(root, fdt_file)
        
        # Construct the corresponding .set file path by searching in the same directory
        set_file_path = os.path.splitext(fdt_file_path)[0] + '.set'
        
        if os.path.exists(set_file_path):
            print(f"Processing file {fdt_file_path} with {set_file_path} in folder {root}")
        
        # Load the EEG data using MNE
            try:
                raw = mne.io.read_raw_eeglab(set_file_path, preload=True)
                
                # Display information about the loaded data
                print(raw.info)

                # Plot the raw data (optional)
                raw.plot()

                # Access the data as a NumPy array for further processing (optional)
                data, times = raw[:]
                
                print(f"Data shape: {data.shape}, Times shape: {times.shape}")
            except Exception as e:
                print(f"Failed to read {set_file_path}: {e}")
        else:
            print(f"Corresponding .set file not found for {fdt_file_path}")

        # Increment the count of files processed
        file_count += 1

# Print the total number of files found
file_count += len(matching_files)
print(f"Total number of files found: {file_count}")
```
